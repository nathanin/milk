```python
import numpy as np
import glob
import os

%matplotlib inline
from matplotlib import pyplot as plt

import pickle
with open('../scripts/dataset/case_dict_obfuscated.pkl', 'rb') as f: 
    case_dict = pickle.load(f)
```

```python
!ls ../scripts/attention/projection/
```

```python
ds = ['2019_02_19_00_55_49',  '2019_02_19_05_09_09',  '2019_02_19_09_33_05']
```

```python
lst = glob.glob('../scripts/attention/projection/2019_02_19_09_33_05/*atns.npy')
print(len(lst))
```

```python
basenames = [os.path.basename(x) for x in lst]
```

```python
!mkdir attention_distributions
```

```python
for basename in basenames:
    plt.clf()
    for d in ds[::-1]:
        caseid = basename.replace('_atns.npy', '')
        ypth = glob.glob(os.path.join('../scripts/attention/projection', d, '{}_*_ys.png'.format(caseid)))[0]
        yhat = float(ypth.split('_')[-2])
        ytrue = case_dict[caseid]
        pth = os.path.join('../scripts/attention/projection', d, basename)
        print(caseid)
        print(ypth)
        print(ytrue)
        print(yhat)
        print(pth)
        
        attns = np.load(pth)
        plt.hist(attns, bins=20, alpha=0.3, label='{} {} {}'.format(d, ytrue, yhat))
        plt.legend(loc=2)
        plt.savefig(os.path.join('attention_distributions', basename.replace('.npy', '.png')))
```

```python
for basename in basenames:
    plt.clf()
    order = None
    for d in ds[::-1]:
        caseid = basename.replace('_atns.npy', '')
        ypth = glob.glob(os.path.join('../scripts/attention/projection', d, '{}_*_ys.png'.format(caseid)))[0]
        yhat = float(ypth.split('_')[-2])
        ytrue = case_dict[caseid]
        pth = os.path.join('../scripts/attention/projection', d, basename)
        print(caseid)
        print(ypth)
        print(ytrue)
        print(yhat)
        print(pth)
        
        attns = np.load(pth)
        if order is None:
            order = np.argsort(attns)
#         plt.hist(attns, bins=20, alpha=0.3, label='{} {}'.format(ytrue, yhat))

        plt.scatter(range(len(attns)), attns[order], alpha=0.3, label='{} {} {}'.format(d, ytrue, yhat))
        plt.legend(loc=2)
        plt.savefig(os.path.join('attention_distributions', basename.replace('.npy', '_sorted.png')))
```
